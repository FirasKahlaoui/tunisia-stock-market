```python
import pandas as pd
```

```python
# read the csv file
df = pd.read_csv('./data/weekly_stock_market.csv')
```

```python
df
```

# Check the DataFrame
*-  if closingPrice < lowerPrice then delete the row*<br>
*-  if openingPrice > highestPrice then delete the row*<br>
*-  if closingPrice > highestPrice then delete the row*<br>
*-  if openingPrice < lowerPrice then delete the row*<br>

```python
df = df[~((df['closingPrice'] < df['lowerPrice']) | 
           (df['openingPrice'] > df['highestPrice']) |
           (df['closingPrice'] > df['highestPrice']) |
           (df['openingPrice'] < df['lowerPrice']))]
df
```
